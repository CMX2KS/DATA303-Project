```python
def clean_text(text: str, year: str = '2020') -> list[str]:
    text = text.split(f'href="https://{year}.igem.org/Team:')
    text = ''.join(text)
    text = text.split('"')
    text = ' '.join(text)
    text = text.split(' ')
    text = [name for name in text if '<' not in name]
    text = [name for name in text if name != '']
    return text
```
