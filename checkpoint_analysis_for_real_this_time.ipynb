```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('IGEM_teams_2004-2024.csv')
data.head(3)
```

```python
years = [year for year in range(2009, 2024)]
tracks = data[data['Year'] >= 2009]['Track'].value_counts().keys()[0:20]
#tracks = [track for track in data[data['Year'] >= 2009]['Track'].unique()]
track_dict = {year:data[data['Year'] == year]['Track'].value_counts().to_dict() for year in years}
```

```python
track_data = pd.DataFrame(data=track_dict)
```

```python
#start figure
plt.figure(figsize=(20,10))

#add stacked bars
for i in range(len(tracks)):
    plt.bar(x=years, height=track_data.loc[tracks[i]].tolist(), color=colors[i%20])

#add titles and labels
plt.title('No. Teams Competing in Each top 20 Tracks by Year')
plt.xlabel('Year')
plt.ylabel('# Teams')
plt.legend(tracks, loc='upper left')

#display the chart
plt.show()
```
